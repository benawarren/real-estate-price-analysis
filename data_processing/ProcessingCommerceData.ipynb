```python
#broader industry classifiers
def classify_industry(description):
    if pd.isna(description):
        return 'Other'
    
    desc = description.lower()

    if any(keyword in desc for keyword in [
        'construction', 'contractor', 'builder', 'flooring', 'demolition', 
        'carpentry', 'roofing', 'drywall', 'remodel', 'paint', 
        'abatement', 'asbestos', 'furnace', 'hvac', 'plumbing', 'fencing']):
        return 'Construction & Trades'

    if any(keyword in desc for keyword in [
        'architectural', 'engineering', 'design']):
        return 'Architecture, Engineering & Design'

    if any(keyword in desc for keyword in [
        'consulting', 'professional service', 'strategic', 'evaluation', 
        'facilitation', 'survey', 'legal', 'community engagement']):
        return 'Consulting & Professional Services'

    if any(keyword in desc for keyword in [
        'education', 'training', 'staffing']):
        return 'Education & Training'

    if any(keyword in desc for keyword in [
        'health', 'wellness', 'counseling', 'medical', 'therapy', 'rehab']):
        return 'Health, Wellness & Medical'

    if any(keyword in desc for keyword in [
        'guard', 'security', 'armed']):
        return 'Security Services'

    if any(keyword in desc for keyword in [
        'retail', 'wholesale', 'supplies', 'building supplies', 
        'textiles', 'uniform']):
        return 'Retail, Wholesale & Supplies'

    if any(keyword in desc for keyword in [
        'marketing', 'advertising']):
        return 'Marketing, Advertising & Media'

    if any(keyword in desc for keyword in [
        'janitorial', 'cleaning', 'facilities management', 'grounds maintenance', 
        'trash', 'recycle', 'laundry']):
        return 'Cleaning & Facility Maintenance'

    if any(keyword in desc for keyword in [
        'real estate', 'property', 'housing']):
        return 'Real Estate & Property Management'

    if any(keyword in desc for keyword in [
        'restaurant', 'catering', 'food']):
        return 'Food Services & Catering'

    if any(keyword in desc for keyword in [
        'technology', 'tech', 'it', 'software', 'digital']):
        return 'Technology & IT'

    if any(keyword in desc for keyword in [
        'trucking', 'transport', 'automotive', 'vehicle', 'delivery', 'logistics']):
        return 'Transportation, Trucking & Automotive'

    if any(keyword in desc for keyword in [
        'finance', 'financial', 'accounting', 'insurance']):
        return 'Finance & Insurance'

    if any(keyword in desc for keyword in [
        'recreation', 'spa', 'beauty', 'barber', 'nails']):
        return 'Recreation & Hospitality'

    return 'Other'

detroit_biz['broad_industry'] = detroit_biz['Industry Type'].apply(classify_industry)
```

```python
#classifying credentials by NAICS code

# Load the CSV file
df = pd.read_csv('Resources/commerce_data/CT_cred_types.csv')

# Ensure the correct column is used
license_types = df['Credential'].dropna().unique()

# Define keyword-based mapping for NAICS categories
naics_mapping = {
    'Other Services (except Public Administration)': [
        'barber', 'beauty', 'laundry', 'massage', 'funeral', 'repair'
    ],
    'Professional, Scientific, and Technical Services': [
        'engineer', 'architect', 'consultant', 'accountant', 'law'
    ],
    'Arts, Entertainment, and Recreation': [
        'entertainment', 'art', 'music', 'performance', 'recreation'
    ],
    'Transportation and Warehousing': [
        'limousine', 'taxi', 'transport', 'trucking', 'delivery', 'warehouse'
    ],
    'Retail Trade': [
        'retail', 'dealer', 'store', 'shop', 'merchandise'
    ],
    'Administrative and Support and Waste Management and Remediation Services': [
        'cleaning', 'janitorial', 'waste', 'pest'
    ],
    'Accommodation and Food Services': [
        'restaurant', 'catering', 'food', 'beverage', 'bakery', 'bar'
    ],
    'Health Care and Social Assistance': [
        'health', 'clinic', 'medical', 'dental', 'care', 'therapy', 'social'
    ],
    'Real Estate and Rental and Leasing': [
        'real estate', 'property', 'leasing', 'rental', 'landlord'
    ]
}

# Create dictionary to hold classified license types
categorized = {category: [] for category in naics_mapping}
categorized['Unclassified'] = []

# Perform classification
for lic in license_types:
    lic_lower = lic.lower()
    matched = False
    for category, keywords in naics_mapping.items():
        if any(keyword in lic_lower for keyword in keywords):
            categorized[category].append(lic)
            matched = True
            break
    if not matched:
        categorized['Unclassified'].append(lic)
```

```python
def convert_credential(cred_type):
    for category in categorized:
        if cred_type in categorized[category]:
            return category
    return 'Unclassified'

just_business['industry_type'] = just_business['Credential'].apply(convert_credential)
just_business['industry_type'].value_counts()
```
